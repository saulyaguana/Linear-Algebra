# Sistemas de Ecuaciones lineales
1. Con una solución
2. Sin solución
3. Infinitas soluciones

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Sin solución

+ Más ecuaciones que variables



```python
x = np.arange(-6, 6)
y = 3 * x + 5
y2 = -x + 3
y3 = 2 * x + 1
```

```python
plt.style.use("ggplot")
fig, axes = plt.subplots(figsize=(8, 6))
axes.plot(x, y)
axes.plot(x, y2)
axes.plot(x, y3)

plt.xlim(-8, 8)
plt.ylim(-8, 8)

axes.axvline(x=0, color="#333")
axes.axhline(y=0, color="#333")

plt.title("Sistema sobredeterminado")

plt.show()
```

# Con solución


```python
x = np.arange(-6, 6)
m = -x + 3
m2 = 2 * x + 1
```

```python
fig, axes = plt.subplots(figsize=(8, 6))
axes.plot(x, m, color='#ED1A57')
axes.plot(x, m2, color='#7A4AED')

axes.axvline(x=0, color='#333')
axes.axhline(y=0, color='#333')

plt.title("Una solución")

plt.show()
```

## Infinitas Soluciones

1. Si me quedo con una ecuación con 2 incógnitas el sistema tiene infinitas soluciones

```python
n = 2 * x + 1
```

```python
fig, axes = plt.subplots(figsize=(8, 6))
axes.plot(x, n)

axes.axvline(x=0, color='#333')
axes.axhline(y=0, color='#333')

plt.title('Infinitas soluciones')

plt.show()
```
