# Combinación Lineal

+ Una combinación lineal es multiplicar un vector por un escalar 2 veces y sumar el resultado de ambos para otener un nuevo vector.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def graficar(vecs, cols, alpha=1):
    plt.figure()
    plt.axvline(x=0, color="#333", zorder=0)
    plt.axhline(y=0, color="#333", zorder=0)
    
    for i in range(len(vecs)):
        x = np.concatenate([[0, 0], vecs[i]])
        plt.quiver([x[0]],
                   [x[1]],
                   [x[2]],
                   [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=cols[i], alpha=alpha)
```

```python
v1 = np.array([2, 5])
v2 = np.array([3, 2])
```

```python
v1v2 = 2 * v1  + 3 * v2
print(v1v2)
```

```python
plt.style.use("Solarize_Light2")

graficar([v1, v2, v1v2], ["orange", "blue", "red"])
plt.xlim(-1, 16)
plt.ylim(-1, 16)
```

```python
for i in range(-10, 10):
    for j in range(-10, 10):
        plt.scatter(v1[0] * i + v2[0] * j, v1[1] * i + v2[1] * j,
                   marker='.',
                   color='orange')
        
plt.xlim(-100, 100)
plt.ylim(-100, 100)
plt.axvline(x=0, color="#333")
plt.axhline(y=0, color="#333")
plt.show()
```
