# Sistema de Ecuaciones Lineales

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5)
y = 3 * x + 5
y2 = 2 * x + 3
```

```python
plt.style.use("Solarize_Light2")
fig, axes = plt.subplots(figsize=(8, 6))
axes.plot(x, y, color="red")
axes.plot(x, y2, color="blue")

axes.axhline(y=0, color="#333")
axes.axvline(x=0, color="#333")

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.show()
```

## Escribiendo el sistema de ecuaciones en forma matricial

+ Para eso hay que dejar el termino intependiente en un lado y del otro solo las variables.





```python
A = np.array([[-3, 1], [-2, 1]])
print(A)
```

```python
B = np.array([[5], [3]])
print(B)
```

```python
sol_1 = np.array([-2, -1])
print(sol_1)
```

```python
A.dot(sol_1)
```

## Ejercicio (práctica)


```python
x = np.arange(-5, 15)
y = 4 - x
y2 = 2 * x - 5
```

```python
fig, axes = plt.subplots(figsize=(8, 6))
axes.plot(x, y, color="blue")
axes.plot(x, y2, color="red")

axes.axvline(x=0, color="#333")
axes.axhline(y=0, color='#333')

plt.xlim(-15, 15)
plt.ylim(-15, 15)

plt.show()
```

```python
A = np.array([[-1, -1], [2, -1]])
B = np.array([[-4], [5]])
```

```python
sol = np.array([[3], [1]])
```

```python
A.dot(sol)
```

```python
A.dot(sol) == B
```
